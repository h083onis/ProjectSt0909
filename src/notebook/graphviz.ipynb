```python
import graphviz
import networkx as nx

# 1. Graphvizでのグラフ生成用のデータ
dot_file_path = 'class_relations.dot'  # 読み込むDotファイルのパス
# target_node = "org.apache.camel.catalog.CamelCatalogTest"  # ターゲットノードを指定
# target_node = "org.apache.camel.catalog.LanguageValidationResult"
# target_node = "org.apache.camel.tooling.model.BaseOptionModel"
# target_node = "org.apache.camle.model.DataFormatModel"
# target_node = "org.apache.camel.catalog.CamelCatalogJSonSchemaResolver"
# target_node = "org.apache.camel.component.undertow.ws.UndertowWsConsumerRouteTest"
# target_node = "org.apache.camel.spi.PackageScanClassResolver"
# target_node = "org.apache.camel.StaticService"
# target_node = "org.apache.camel.parser.xml.XmlOnExceptionRouteTest"
target_node = "org.apache.camel.ApiEndpoint"

# 2. networkxでDotファイルを読み込む
G = nx.DiGraph(nx.drawing.nx_pydot.read_dot(dot_file_path))

# 3. target_nodeに関連するノードを収集
descendants = nx.descendants(G, target_node)  # 呼び出されている側（子ノード）
# ancestors = nx.ancestors(G, target_node)      # 呼び出している側（親ノード）

# 自分自身もセットに追加
# related_nodes = descendants | ancestors
related_nodes = descendants
# related_nodes = ancestors
related_nodes.add(target_node)

# 4. 新しいGraphvizグラフを作成
subgraph = graphviz.Digraph(format='png')
subgraph.attr('graph', rankdir='LR')  # 左から右のレイアウトを指定（任意）

# 5. 収集したノードと関連エッジだけを追加
for node in related_nodes:
    subgraph.node(node)

for u, v in G.edges():
    if u in related_nodes and v in related_nodes:
        subgraph.edge(u, v)

# 6. 画像として保存
# output_path = 'filtered_graph_bidirectional5'
output_path = target_node
subgraph.render(filename=output_path, cleanup=True)

print(f"{output_path}.png に画像を保存しました。")
```

```python



# Testファイルを再帰的に探す関数
def find_test_file(graph, start_node):
    test_files = []
    stack = [(start_node, 0)]  # 探索用スタック
    visited = set()  # 訪問済みノードの記録
    while stack:
        current_node, depth = stack.pop()
        if current_node in visited:
            continue
        visited.add(current_node)
        
        # ノード名に"Test"が含まれている場合
        if "Test" in current_node:
            test_files.append((current_node, depth))
        for predecessor in graph.predecessors(current_node):
            stack.append((predecessor, depth + 1))
    
    return test_files

# 起点ノード
start_node = "org.apache.camel.tooling.model.ReleaseModel"

# Testファイルを探索
test_file = find_test_file(G, start_node)

if test_file:
    print(f"見つかったTestファイル: {test_file}")
    for (test,depth) in test_file:
        if depth == 1:
            print(test, depth)
else:
    print("Testファイルが見つかりませんでした。")
```
