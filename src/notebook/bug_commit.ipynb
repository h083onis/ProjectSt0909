```python
# ãƒ•ã‚£ãƒ«ã‚¿ãƒªãƒ³ã‚°å‡¦ç†
filtered_dfs = []

for key,values in data2.items():
    filtered_df = result_df[
        (result_df["commit_id"] == key) & 
        (result_df["changed_src_files"].isin(values))
    ]
    # print(filtered_df)
    filtered_dfs.append(filtered_df)
    

# å…¨ã¦ã‚’çµåˆï¼ˆå¿…è¦ãªã‚‰ï¼‰
result_df2 = pd.concat(filtered_dfs, ignore_index=True)
```
